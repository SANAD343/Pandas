# Descriptive statistics

```python
import numpy as np
data={"a":np.arange(11,21),"b":np.arange(21,31),"c":np.arange(31,41)} #numpy is more efficeient then pandas
df=pd.DataFrame(data)
print(df)
```

```python
df.sum()
```

```python
df.sum(axis=1)
```

```python
df.mean()
```

```python
df.mean(skipna=True) # this command is use when we have NaN data present in the array
```

```python
df.min()
```

```python
df.var() # Variance
```

```python
df.max() # Max Value
```

```python
df.std() # Standard Deviation
```

```python
df.median()
```

```python
df.mode()
```

```python
df.describe() # Generates descriptive statistics that summarize the central tendency, dispersion and shape of a dataset's distribution, excluding NaN values
```
